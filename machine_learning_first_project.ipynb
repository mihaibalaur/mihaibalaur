# **0 - Data Explanation**

- dependent variable / output value

## *X* - represented by the other columns of the dataset
- independent variable / input value

# **1 - Data Loading - `.csv` data format**

```python
import pandas as pd


DATASET = 'https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv'

# df - dataframe
df = pd.read_csv(DATASET)

# print the dataframe content
df
```

# **2 - Data Preparation**

### **2.1 - Data Separation as `X` and `y`**

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

### **2.2 - Data Splitting - Training Set | Testing Set**

```python
from sklearn.model_selection import train_test_split


# test_size - the proportion of the dataset to include in the TEST SPLIT
# random_state - controls the shuffling applied to the data before applying the split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

# 915 rows = 0.8 x 1144 total rows
# data used to train the model
X_train
```

```python
# 229 rows = 0.2 x 1144 total rows
# data used to test the model
X_test
```

# **3 - Model Building**

## **3.1 - Linear Regression**

### **3.1.1 - Training the Model**

```python
from sklearn.linear_model import LinearRegression


# instantiate the linear regression model
lr = LinearRegression()

# fit - means finding the best straight line (y = mx + b) that minimizes the total distance (errors/residuals) between the line and the actual data points

# the model is trained on the `X_train` dataset to predict the `y_train` dataset
lr.fit(X_train, y_train)
```

### **3.1.2 - Testing the Model - Applying the Model to make a Prediction**

```python
# it will predict based upon the original dataset - `X_train` - that it has been trained on in order to evaluate the performance of the algorithm

# the output should be similar to `y_train`, not identical
# identical output => OVERFITTING - not good
y_lr_train_pred = lr.predict(X_train)

# make a new prediction on a new dataset
y_lr_test_pred = lr.predict(X_test)


y_lr_train_pred
```

```python
y_lr_test_pred
```

### **3.1.3 - Model Performance Evaluation**

```python
from sklearn.metrics import mean_squared_error, r2_score


# The performance is established by comparing `y_train` - TRUE VALUES and `y_lr_train_pred` - PREDICTED VALUES

# MSE/R2S - evaluation metrics for regression models
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

### **3.1.4 - Model Performance Comparison DataFrame**

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Testing MSE', 'Testing R2']

lr_results
```

## **3.2 - Random Forest**

### **3.2.1 - Training the Model**

```python
from sklearn.ensemble import RandomForestRegressor


# instantiate the random forest model
rf = RandomForestRegressor(max_depth=2, random_state=100)

rf.fit(X_train, y_train)
```

### **3.2.2 - Testing the Model - Applying the Model to make a Prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **3.2.3 - Model Performance Evaluation**

```python
from sklearn.metrics import mean_squared_error, r2_score


# The performance is established by comparing `y_train` - TRUE VALUES and `y_lr_train_pred` - PREDICTED VALUES

# MSE/R2S - evaluation metrics for regression models
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

### **3.2.4 - Model Performance Comparison DataFrame**

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Testing MSE', 'Testing R2']

rf_results
```

# **4 - Models Performance Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models.reset_index(drop=True)
```
